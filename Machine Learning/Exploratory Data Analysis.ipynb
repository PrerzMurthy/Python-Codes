<b>6. Number of new cases plot

```python
new_cases = data.loc[:, ['cancer_type', 'new_cases_number']]
new_cases=new_cases[new_cases.new_cases_number>20000]
print(new_cases)
```

```python
plt.figure(figsize=(15,10))
plt.title("2020 New Cancer Cases In India")
new_cases=new_cases.sort_values(['new_cases_number'])
sns.barplot(x=new_cases['cancer_type'], y=new_cases['new_cases_number'])
plt.xticks(rotation= 90)
plt.xlabel("Cancer Type")
plt.ylabel("Number of New Cases")
```

<b>7. Getting top 5 types of cancer in terms of new cases, risk, death and plotting them

```python
data.sort_values(by=['new_cases_number'],ascending=False)['cancer_type'].head(5)
```

```python
df1 = (data.sort_values(by=['new_cases_number'],ascending=False)['cancer_type'].head(5)).index.values
sorted_data = data.reindex(df1)
plt.figure(figsize=(15,10))
sns.barplot(x=sorted_data['cancer_type'],y=sorted_data['new_cases_number'])
```

```python
data.sort_values(by=['Deaths Number'],ascending=False)['cancer_type'].head(5)
```

```python
df2 = (data.sort_values(by=['Deaths Number'],ascending=False)['cancer_type'].head(5)).index.values
sorted_data = data.reindex(df2)
plt.figure(figsize=(15,10))
sns.barplot(x=sorted_data['cancer_type'],y=sorted_data['Deaths Number'])
```

```python
data.rename(columns={'per100k_5yr_prevalence':'Risk'},inplace=True)
data.dtypes
data.sort_values(by=['Risk'],ascending=False)['cancer_type'].head(5)
```

```python
df3 = (data.sort_values(by=['Risk'],ascending=False)['cancer_type'].head(5)).index.values
sorted_data = data.reindex(df3)
plt.figure(figsize=(15,10))
sns.barplot(x=sorted_data['cancer_type'],y=sorted_data['Risk'])
```
