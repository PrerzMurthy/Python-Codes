<font color="blue"><font size="4"><b>2. Fetch all the rows having Salary >80000

```python
#Fetching the rows having salary > 80000
salary = employee[employee['Salary'] > 80000]
salary
```

<font color="blue"><font size="4"><b>3. Fetch all the rows having Team = Legal and Finance.

```python
#Fetching the rows having Team = Legal and Finance
LF = employee[employee['Team'].isin(['Finance', 'Legal']) ]
LF
```

<font color="blue"><font size="4"><b>4. Find the missing values in Senior Management and replace it with NONE.

```python
#Finding the missing values in Senior Management and replace it with NONE
employee['Senior Management'] = employee['Senior Management'].replace(np.nan, 'NONE')
employee
```

<font color="blue"><font size="4"><b>5. Find the employees of having Bonus % more than 12 and reduce it with 2.

```python
#Finding the employees of having Bonus % more than 12 and reduce it with 2.
employee.loc[employee.Bonus >= 12, "Bonus"] = employee.Bonus - 2
employee
```
