<font face="Bookman Old Style"><font size="4"><b>1. Create a lambda function to implement equations for the following financial tasks:
<font size="3"><font color="lavendar"><center>i) Loss<br>
ii) Profit<br>
iii) Simple interest<br>
    iv) Compound interest</b></center>

```python
loss = lambda cp,sp: cp-sp
profit = lambda cp,sp: sp-cp
s_interest = lambda p,r,t: (p*t*r)/100
c_interest = lambda p,r,t,n: (p*((1+(r/100))**(n*t)))-p

ch = int(input("What do you wanna calculate? \n1:Loss\n2:Profit\n3:Simple interest\n4:Compound interest\n"))
if ch==1:
    cp = float(input('\nEnter the original price of the item (Cost price): '))
    sp = float(input('Enter the price at which the item was sold (Selling price): '))
    if loss(cp,sp)<0:
        print('\nResult:\nThere is no loss in this purchase.')
        print('The profit in this purchase = Rs.',abs(loss(cp,sp)))
    else:
        print('\nResult:\nThe loss in this purchase = Rs.',loss(cp,sp))
        
elif ch==2:
    cp = float(input('\nEnter the original price of the item (Cost price): '))
    sp = float(input('Enter the price at which the item was sold (Selling price): '))
    if profit(cp,sp)<0:
        print('\nResult:\nThere is no profit in this purchase.')
        print('The loss in this purchase = Rs.',abs(profit(cp,sp)))
    else:
        print('\nResult:\nThe profit in this purchase = Rs.',profit(cp,sp))
        
elif ch==3:
    p = float(input('\nEnter the principal amount: '))
    r = float(input('Enter the annual interest rate (in percentage): '))
    t = float(input('Enter the time period (in years): '))
    print('\nResult:\nSimple Interest = Rs.',s_interest(p,r,t))

elif ch==4:
    p = float(input('\nEnter the principal amount: '))
    r = float(input('Enter the annual interest rate (in percentage): '))
    t = float(input('Enter the number of time periods elapsed (in years): '))
    n = float(input('Enter the number of times the interest is compounded per year: '))
    print('\nResult:\nCompound Interest = Rs.',c_interest(p,r,t,n))
```

```python
import numpy as np
from abc import ABC, abstractmethod

class Shapes(ABC):
    @abstractmethod
    def area_vol(self):
        pass

class Sphere(Shapes):
    def area_vol(self):
        r = float(input('Enter the radius of the sphere: '))
        ssphere = 4*np.pi*(r**2)
        vsphere = (4/3)*np.pi*(r**3)
        print('\nResult:\nSurface area of the sphere =',ssphere)
        print('Volume of the sphere =',vsphere)
                
class Cylinder(Shapes):
    def area_vol(self):
        r = float(input('Enter the radius of the cylinder: '))
        h = float(input('Enter the height of the cylinder: '))
        scyl = 2*np.pi*r*(r+h)
        vcyl = np.pi*(r**2)*h
        print('\nResult:\nSurface area of the cylinder =',scyl)
        print('Volume of the cylinder =',vcyl)     
        
class Cone(Shapes):
    def area_vol(self):
        r = float(input('Enter the radius of the cone: '))
        s = float(input('Enter the slant length of the cone: '))
        h = float(input('Enter the height of the cone: '))
        scone = np.pi*r*(s+r)
        vcone = (1/3)*np.pi*(r**2)*h
        print('\nResult:\nSurface area of the cone =',scone)
        print('Volume of the cone =',vcone)       
        
class Rectangular_Prism(Shapes):
    def area_vol(self):
        l = float(input('Enter the length of rectangular prism: '))
        w = float(input('Enter its width: '))
        h = float(input('Enter its height: '))
        srp = 2*(l*w+l*h+w*h)
        vrp = l*w*h
        print('\nResult:\nSurface area of the rectangular prism =',srp)
        print('Volume of the rectangular prism =',vrp)     
        
class Triangular_Prism(Shapes):
    def area_vol(self):
        b = float(input('Enter length of base of triangle: '))
        h = float(input('Enter the height of the prism: '))
        l = float(input('Enter the length of the rectangular part of prism: '))
        s = float(input('Enter the side of triangle: '))
        stp = b*h+l*b+2*l*s
        vtp = (1/2)*b*l*h
        print('\nResult:\nSurface area of the triangular prism =',stp)
        print('Volume of the triangular prism =',vtp)        
        

print('\nProgram to find the surface area and volume of shapes:')
print('______________________________________________________\n')
print('MENU: ')
print('----\n')
menu = {1:'Sphere',2:'Cylinder',3:'Cone',4:'Rectangular Prism',5:'Triangular Prism'}
for x,y in menu.items():
    print('{0}. {1}'.format(x,y))

# Objects for each class is created:
S = Sphere()
Cy = Cylinder()
Co = Cone()
R = Rectangular_Prism()
T = Triangular_Prism()

try:
    n = 0
    while n<=0 or n>5:
        n = int(input('\nEnter your choice: '))
    
    switcher = {
        1: S.area_vol,
        2: Cy.area_vol,
        3: Co.area_vol,
        4: R.area_vol,
        5: T.area_vol
        }   
    
    switcher[n]()
    
except ValueError:
    print('\nInvalid input!\nEnd of the program.')
```
