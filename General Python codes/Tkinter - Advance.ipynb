```python
from openpyxl import *
import tkinter
from tkinter import *
from tkinter import ttk
from tkinter import scrolledtext
from tkinter.ttk import *
from tkinter.ttk import Progressbar
from PIL import Image,ImageTk 
import time
import tkinter as tk
import tkinter.messagebox
  
wb = load_workbook('C:\\Users\\Prerana Murthy\\Documents\\Christ\\SEM 5\\Python programming lab\\reading_rooms_customers.xlsx')
sheet = wb.active

def new_reg_win():
    
    reg_win = tk.Toplevel(main_win)
    reg_win.title("Reading rooms")
    reg_win.geometry("3000x3000")
    reg_win.configure(bg='navy')
       
    def excel():
        sheet.column_dimensions['A'].width = 30
        sheet.column_dimensions['B'].width = 10
        sheet.column_dimensions['C'].width = 10
        sheet.column_dimensions['D'].width = 60
        sheet.column_dimensions['E'].width = 20
        sheet.column_dimensions['F'].width = 40
        sheet.column_dimensions['G'].width = 40
        sheet.column_dimensions['H'].width = 50

        sheet.cell(row=1, column=1).value = "Name"
        sheet.cell(row=1, column=2).value = "Age"
        sheet.cell(row=1, column=3).value = "Gender"
        sheet.cell(row=1, column=4).value = "Address"
        sheet.cell(row=1, column=5).value = "Email ID"
        sheet.cell(row=1, column=6).value = "Phone number"
        sheet.cell(row=1, column=7).value = "Field of study"
        sheet.cell(row=1, column=8).value = "Interested languages"

    def focus1(event):
        age_field.focus_set()

    def focus2(event):
        gender_field.focus_set()

    def focus3(event):
        address_field.focus_set()

    def focus4(event):
        email_id_field.focus_set()

    def focus5(event):
        phone_no_field.focus_set()

    def focus6(event):
        study_field.focus_set()
    
    def focus7(event):
        lang_field.focus_set()
        
    def about():
        tkinter.messagebox.showinfo("What is it about?","Please fill in all the details and you will be registered with Reading rooms!")
        reg_win.lift()
        
    def step():
        ph_str = phone_no_field.get()
        str(ph_str)
        fail=1
        if len(ph_str) != 10 or ph_str[0] == "0" or ph_str.isalpha():
            tkinter.messagebox.showwarning("Warning","Invalid Mobile number")
            fail=0
        if fail==1:
            for i in range(10):
                reg_win.update_idletasks()
                pb1['value'] += 10
                time.sleep(0.2)
            lb11 = Label(reg_win,text="")
            lb11.configure(background="navy")
            lb11.place(height=30, width=250, x=500, y=570)
            tkinter.messagebox.showinfo("Thank you","Thank you registering with us!")
            insert()
        else:
            phone_no_field.delete(0, END)
            reg_win.lift()

    #Clear the data after it is saved into excel
    def clear():
        name1_field.delete(0, END)
        age_field.delete(0, END)
        gender_field.set('')
        address_field.delete(1.0, END)
        email_id_field.delete(0, END)
        phone_no_field.delete(0, END)
        study_field.delete(0, END)
        lang_field.delete(0,END)
        lb10 = Label(reg_win, text="Click the save button!",font=("Arial bold",20))
        lb10.configure(foreground="deeppink",background="navy")
        lb10.place(width=500, x=485, y=520)
        pb1 = Progressbar(reg_win, orient=HORIZONTAL, length=100, mode='determinate')
        pb1.place(width=250, x=500, y=560)

    #Insert data from GUI to excel
    def insert():
        if (name1_field.get() == "" and
            age_field.get() == "" and
            gender_field.get() == "" and
            address_field.get() == "" and
            email_id_field.get() == "" and
            phone_no_field.get() == "" and
            study_field.get() == ""and
            lang_field.get() == ""):
            print("---")
        else:
            current_row = sheet.max_row
            current_column = sheet.max_column
            sheet.cell(row=current_row + 1, column=1).value = name1_field.get()
            sheet.cell(row=current_row + 1, column=2).value = age_field.get()
            sheet.cell(row=current_row + 1, column=3).value = gender_field.get()
            sheet.cell(row=current_row + 1, column=4).value = address_field.get(1.0, END)
            sheet.cell(row=current_row + 1, column=5).value = email_id_field.get()
            sheet.cell(row=current_row + 1, column=6).value = phone_no_field.get()
            sheet.cell(row=current_row + 1, column=7).value = study_field.get()
            sheet.cell(row=current_row + 1, column=8).value = lang_field.get()
            wb.save('C:\\Users\\Prerana Murthy\\Documents\\Christ\\SEM 5\\Python programming lab\\reading_rooms_customers.xlsx')
            name1_field.focus_set()
            # call the clear() function
            clear()
    
    excel()

    lb1 = Label(reg_win, text="REGISTRATION FORM", font=("Verdana Bold",25))
    lb1.configure(foreground="deeppink",background="navy")
    lb1.place(width=450, x=450, y=30)

    lb2 = Label(reg_win, text="Full Name: ",font=("Arial Bold",15))
    lb2.configure(foreground="deeppink",background="navy")
    lb2.place(width=150, x=20, y=100)

    name1_field = Entry(reg_win,width=50)
    name1_field.place(width=150, x=250, y=100)

    lb3 = Label(reg_win, text="Age: ",font=("Arial Bold",15))
    lb3.configure(foreground="deeppink",background="navy")
    lb3.place(width=100, x=20, y=150)

    age_field = Spinbox(reg_win, from_ =10, to=100, width=5)
    age_field.place(width=100, x=250, y=150)

    lb4 = Label(reg_win, text="Gender: ",font=("Arial Bold",15))
    lb4.configure(foreground="deeppink",background="navy")
    lb4.place(width=100, x=20, y=200)

    gender_field=IntVar()
    rad1 = Radiobutton(reg_win,text="Male",value=1, variable=gender_field)
    rad2 = Radiobutton(reg_win,text="Female",value=2, variable=gender_field)
    rad1.place(width=100, x=250, y=200)
    rad2.place(width=100, x=450, y=200)

    def clicked():
        print(gender_field.get())

    lb5 = Label(reg_win, text="Address: ",font=("Arial Bold",15))
    lb5.configure(foreground="deeppink",background="navy")
    lb5.place(width=100, x=20, y=250)

    address_field = scrolledtext.ScrolledText(reg_win,width=100,height=2)
    address_field.place(width=200, x=250, y=250)

    lb6 = Label(reg_win, text="Email ID: ",font=("Arial Bold",15))
    lb6.configure(foreground="deeppink",background="navy")
    lb6.place(width=100, x=20, y=350)

    email_id_field = Entry(reg_win,width=50)
    email_id_field.place(width=100, x=250, y=350)

    lb7 = Label(reg_win, text="Phone Number: ",font=("Arial Bold",15))
    lb7.configure(foreground="deeppink",background="navy")
    lb7.place(width=200, x=20, y=400)

    phone_no_field = Entry(reg_win,width=20,justify = LEFT)
    phone_no_field.place(width=100, x=250, y=400)

    lb8 = Label(reg_win, text="Field of Study: ",font=("Arial Bold",15))
    lb8.configure(foreground="deeppink",background="navy")
    lb8.place(width=400, x=20, y=450)

    study_field = Combobox(reg_win)
    study_field['values'] = ("Business/Commerce","Computer Science/Technology/Engineering","Humanities/Arts","Law","MBA/PhD","Other")
    study_field.current(5)
    study_field.place(width=100, x=250, y=450)
    
    lb9 = Label(reg_win, text="Interested languages: ",font=("Arial Bold",15))
    lb9.configure(foreground="deeppink",background="navy")
    lb9.place(width=530, x=20, y=500)
    
    lang_field = Entry(reg_win,width=100)
    lang_field.place(width=150, x=250, y=500)

    excel()

    menubar = Menu(reg_win)  
    file = Menu(menubar, tearoff=0)  
    file.add_command(label="New")  
    file.add_command(label="Open")  
    file.add_command(label="Save")  
    file.add_command(label="Save as")    
    file.add_separator()  
    file.add_command(label="Exit")  
    menubar.add_cascade(label="File", menu=file)  

    edit = Menu(menubar, tearoff=0)  
    edit.add_command(label="Undo")  
    edit.add_separator()     
    edit.add_command(label="Cut")  
    edit.add_command(label="Copy")  
    edit.add_command(label="Paste")  
    menubar.add_cascade(label="Edit", menu=edit)  

    help = Menu(menubar, tearoff=0)  
    help.add_command(label="About", command=about)  
    menubar.add_cascade(label="Help", menu=help)  

    reg_win.config(menu=menubar)

    lb10 = Label(reg_win, text="Click the save button!",font=("Arial bold",20))
    lb10.configure(foreground="deeppink",background="navy")
    lb10.place(width=500, x=485, y=520)

    pb1 = Progressbar(reg_win, orient=HORIZONTAL, length=100, mode='determinate')
    pb1.place(width=250, x=500, y=560)

    button=tk.Button(reg_win, text= "Save",font=("Arial Bold",15),bg="cyan",fg="deeppink",command= step)
    button.configure(relief=SUNKEN)
    button.place(width=100, x=580, y=590)

if __name__ == "__main__":
    
    main_win = tk.Tk()
    main_win.title("Welcome page")
    main_win.geometry("3000x3000")
    main_win.configure(bg='navy')
    image1 = Image.open("C:\\Users\\Prerana Murthy\\Documents\\Christ\\SEM 5\\Python programming lab\\pic4.jpg")
    resize_image = image1.resize((100,75),Image.ANTIALIAS)
    test = ImageTk.PhotoImage(resize_image)

    label1 = tkinter.Label(image=test)
    label1.image = test
    label1.place(width=100, x=30, y=30)

    image2 = Image.open("C:\\Users\\Prerana Murthy\\Documents\\Christ\\SEM 5\\Python programming lab\\pic3.jpg")
    resize_image = image2.resize((100,75),Image.ANTIALIAS)
    test = ImageTk.PhotoImage(resize_image)

    label2 = tkinter.Label(image=test)
    label2.image = test
    label2.place(width=100, x=1150, y=30)

    l = Label(main_win, text = "Welcome to Reading rooms!")
    l.config(font =("Verdana", 30))
    l.configure(foreground="deeppink",background="navy")

    T = Text(main_win, height = 12, width = 150)
    T.configure(bg="navy",fg="deeppink")
    T.config(font=("Arial Bold",15))
    T.tag_configure("Fact",justify='center')

    Fact = """Everyone at Reading rooms believes in the importance of books in our lives and how they can shape and influence us. 
    \nAll of us wish to promote book reading habits across all age-groups and more importantly in the next-gen children. 
    \nWe want to be the reading face of the new generation. We have a very good collection of more than 14000 books. And we are always on the prowl, adding books every month. 
    \nWe store books in regional languages (Tamil, Telugu, Malayalam, Hindi, Bengali, Kannada and so many more). We add all the popular,\nbest-seller and critically acclaimed books immediately on release. We also have books related to various levels of \neducation. \n                                            REGISTER BELOW TO BECOME A MEMBER OF READING ROOMS!!"""

    l.place(width=600,x=350,y=100)
    T.place(width=1200,x=35,y=250)
    T.insert(tk.END, Fact)

    button = tk.Button(main_win, text="Register here", bg='black', fg='cyan',font=("Arial Bold",10),
                                  command=lambda: new_reg_win())
    button.configure(relief=RAISED)

    button.place(width=200,x=550,y=600)
    reg_win.mainloop()
    
```
