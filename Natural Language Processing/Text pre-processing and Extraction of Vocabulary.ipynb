```python
def extract_animal_names(text):
    tokens = nltk.word_tokenize(text.lower())
    return [word for word in tokens if word in animals]
```

```python
story1_animals = set(extract_animal_names(new_text1))
story1_animals
```

```python
story2_animals = set(extract_animal_names(new_text2))
story2_animals
```

```python
common_animals = story1_animals.intersection(story2_animals)
if common_animals:
    print("The two stories talk about:", ', '.join(common_animals))
else:
    print("There are no animals common between the two stories.")
```

<font size="6"><font face="Bookman Old Style"><b><center>Question 2</center></b><br>
<font size="5"><b>Develop python code to help a user buy a computer online:</b><br>
<font size="3">a) Create a list with product specifications of minimum 5 computers as follows:
('Lenovo 5GHz 600GB 900', 'Toshiba 6GHz 800GB 990', 'Dell 8GHz 500GB 980', 'HP 8Ghz 500GB 1100', 'Asus 6GHz 400GB 900')<br>
b) The user should give his requirement as a sentence for example: “computer with
at least 6GHz and at least 500GB storage for not more than 1000”. Using regular
expression matching find the specifications like 6GHz, 500GB and 1000.<br>
c) Using regular expression matching find the computers that satisfy the user’s
requirements. If there are products in the list that match user’s criteria:
- Then print: There are <count> computers available: <Display all the matching
product strings in new lines>
- Else print: There are no computers available matching your requirements!

```python
computer_specs = [
    'Lenovo 5GHz 600GB 900',
    'Toshiba 6GHz 800GB 990',
    'Dell 8GHz 500GB 980',
    'HP 8GHz 500GB 1100',
    'Asus 6GHz 400GB 900'
]
computer_specs
```

```python
import re

user_input = input("Please enter your requirements: ")
pattern = r'(\d+)GHz.*?(\d+)GB.*?(\d+)'

match = re.search(pattern, user_input)
if match:
    min_clock_speed = int(match.group(1))
    min_storage = int(match.group(2))
    max_price = int(match.group(3))
    matched_text = match.group(0)
    print("Matched text:", matched_text)
else:
    print("Invalid input format.")
```

```python
matching_computers = []
for specs in computer_specs:
    spec_match = re.search(pattern, specs)
    if spec_match:
        comp_clock_speed = int(spec_match.group(1))
        comp_storage = int(spec_match.group(2))
        comp_price = int(spec_match.group(3))
        
        if comp_clock_speed >= min_clock_speed and comp_storage >= min_storage and comp_price <= max_price:
            matching_computers.append(specs)

if matching_computers:
    print("There are computers available:")
    for comp in matching_computers:
        print(comp)
else:
    print("There are no computers available matching your requirements!")
```
