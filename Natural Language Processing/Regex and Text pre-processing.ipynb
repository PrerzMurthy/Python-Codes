<font size="3">a) Read the content of the book &#39;Indian Fairy Tales&#39; from the gutenberg_book.txt file.

```python
with open("gutenberg_book.txt",'r') as file:
  text = file.read()
  print(text)
```

<font size="5"><b>2. Regular expressions:</b>

<font size="3">a) Extract the Title, Editor, Release Date, Most recently update, Language and Character set
encoding from the book.

```python
import re

title_pattern = r"Title: (.+)"
editor_pattern = r"Editor: (.+)"
release_date_pattern = r"Release Date: (.+)"
recent_update_pattern = r"Most recently updated: (.+)"
language_pattern = r"Language: (.+)"
charset_pattern = r"Character set encoding: (.+)"

title = re.search(title_pattern, text).group(1)
editor = re.search(editor_pattern, text).group(1)
release_date = re.search(release_date_pattern, text).group(1)
recent_update = re.search(recent_update_pattern, text).group(1)
language = re.search(language_pattern, text).group(1)
charset_encoding = re.search(charset_pattern, text).group(1)

# Print the extracted information
print("Title:", title)
print("Editor:", editor)
print("Release Date:", release_date)
print("Most Recently Update:", recent_update)
print("Language:", language)
print("Character Set Encoding:", charset_encoding)
```

b) Extract all email ids.

```python
email_pattern = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b'
email_ids = re.findall(email_pattern, text)
email_ids
```

c) Extract usernames from email ids.

```python
usernames = [re.search(r"([A-Za-z0-9._%+-]+)@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}", email).group(1) for email in email_ids]
usernames
```

d) Replace email ids with the email id &#39;support@christuniversity.in&#39;.

```python
replaced_text = re.sub(email_pattern, 'support@christuniversity.in', text)
replaced_text
```

e) Extract all website URLs.

```python
url_pattern = r'\bhttps?://[^\s<>"]+|www\.[^\s<>"]+\b'
urls = re.findall(url_pattern, text)
urls
```

f) Extract all 9-lettered words in the text and print the list of those words.

```python
nine= re.findall(r'\b\w{9}\b', text)
nine
```

<font size="5"><b>3. Text Processing:</b>


<font size="3">a) Lowercase the book content and write it into a file.

```python
import nltk
```

```python
normalised_text = text.lower()

output_file_path = "lower_CAC.txt"

with open("lower_CAC.txt",'w') as file:
  for sentence in normalised_text:
    file.write(sentence)

print("Results written to", output_file_path)
```

b) Remove all punctuation from the book. Print the list of punctuation that were removed.

```python
with open("lower_CAC.txt",'r') as file:
  text_1 = file.read()

from nltk.tokenize import word_tokenize
import string
import re
nltk.download('punkt')

tokens = word_tokenize(text_1)
pattern = r'[^\w\s]'
new_text1 = re.sub(pattern, '', text_1)

punc_removed = re.findall(pattern, text_1)

output_file_path = "punc.txt"

with open('punc.txt', 'w') as file:
    file.write(new_text1+'\n\n')
    file.write(str(punc_removed))

print("Results written to", output_file_path)
```
