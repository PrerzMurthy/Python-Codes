<font size="4"><font face="Bookman Old Style"><b>Question 2</b><br>

```python
import nltk
from nltk.corpus import wordnet
```

```python
word = "old"
synsets = wordnet.synsets(word)

for synset in synsets:
    print("Definition:", synset.definition())
    antonyms = []
    for lemma in synset.lemmas():
        for antonym in lemma.antonyms():
            antonyms.append(antonym.name())
    print("Antonyms:", antonyms)
    print()
```

<font size="4"><font face="Bookman Old Style"><b>Question 3</b><br>

```python
ab_list = ['lion', 'hippo', 'rhino', 'gazelle', 'monkey', 'chicken', 'ostrich', 'vulture', 'wolf', 'hyena']

for word in ab_list:
    synsets = wordnet.synsets(word)
    if synsets:
        hypernyms = []
        for synset in synsets:
            for hypernym in synset.hypernyms():
                hypernyms.append(hypernym.lemmas()[0].name())
        print(f"{word}: {', '.join(hypernyms)}")
    else:
        print(f"No synsets found for {word}")
```

<font size="4"><font face="Bookman Old Style"><b>Question 4</b><br>

```python
animal_bird_list = ['lion', 'hippo', 'rhino', 'gazelle', 'monkey', 'chicken', 'ostrich', 'vulture', 'wolf', 'hyena']
target_word = 'hyena'

target_synset = wordnet.synsets(target_word)[0]

most_similar_word = None
highest_similarity = -1

for word in animal_bird_list:
    word_synset = wordnet.synsets(word)[0]
    similarity = target_synset.wup_similarity(word_synset)
    if similarity > highest_similarity:
        highest_similarity = similarity
        most_similar_word = word

print(f"The most similar animal/bird to '{target_word}' is '{most_similar_word}' with similarity: {highest_similarity:.4f}")
```

```python
import nltk
from nltk import FreqDist
from nltk.corpus import stopwords

nltk.download('stopwords')
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download('punkt')
nltk.download('stopwords')

# Tokenize the story into words
words = word_tokenize(story.lower())  # Tokenize and convert to lowercase

# Remove stopwords
filtered_words = [word for word in words if word not in stopwords.words('english')]

# Reconstruct the story without stopwords
filtered_story = ' '.join(filtered_words)

print(filtered_story)
```

```python
token1 = word_tokenize(filtered_story)
list(token1)
```

```python
import pandas as pd

fd = FreqDist(token1)
import sys
initial_max_display = pd.options.display.max_rows
pd.options.display.max_rows = len(fd)

# Display the FreqDist object
print(fd)

word_frequency_list = [(word, frequency) for word, frequency in fd.items()]
print(word_frequency_list)
```

<font size="4"><font face="Bookman Old Style"><b>Question 7</b><br>

```python
import matplotlib.pyplot as plt

# Create a final list of words with frequency greater than 3
final_word_list = []
for word, frequency in word_frequency_list:
    if frequency > 3:
        final_word_list.append(word)

# Plot frequency distribution
freq_dist = {word: frequency for word, frequency in word_frequency_list}
plt.figure(figsize=(35, 6))
plt.bar(freq_dist.keys(), freq_dist.values())
plt.xticks(rotation=90)
plt.title("Words with Frequency Greater than 3")
plt.xlabel("Words")
plt.ylabel("Frequency")
plt.show()
```

<font size="4"><font face="Bookman Old Style"><b>Question 8</b><br>

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Create a dictionary from the word_frequency_list
word_freq_dict = {word: frequency for word, frequency in word_frequency_list}

# Create the WordCloud object
wordcloud = WordCloud(width=800, height=400, background_color='white', max_words=100).generate_from_frequencies(word_freq_dict)
```

```python
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
